Deep Neural Network Model 

```
def R2_coefficient(y_true, y_pred):
    RSS =  K.sum(K.square( y_true-y_pred ))
    TSS = K.sum(K.square( y_true - K.mean(y_true) ) )
    return ( 1 - RSS/(TSS + K.epsilon()) )
```

data preparation

```
X_train_np = X_train.to_numpy()
y_train_np = y_train.to_numpy()
X_test_np = X_test.to_numpy()
y_test_np = y_test.to_numpy()
print(X_train_np.shape, y_train_np.shape, X_test_np.shape, y_test_np.shape)
```

building DNN model

```
def model():
  inp = keras.layers.Input(shape = (None, 17))
  out = tf.keras.layers.Dense(30, activation = 'elu')(inp)
  out = tf.keras.layers.Dense(20, activation = 'sigmoid')(out)
  out = tf.keras.layers.Dense(15, activation = 'sigmoid')(out)
  out = tf.keras.layers.Dense(10, activation = 'sigmoid')(out)
  out = tf.keras.layers.Dense(1)(out)
  mod = tf.keras.Model(inp, out)
  return mod
```

```
mod = model()
mod.summary()
```

```
optimiser = tf.keras.optimizers.Adam(learning_rate=0.01)
loss = 'mse'
mod.compile(optimizer=optimiser, loss=loss, metrics=[R2_coefficient, 'mae', 'mse'])
```

training the model

```
mod.fit(
    x=X_train_np, y=y_train_np, batch_size=12711-1272, epochs=10000, verbose=1, callbacks=None,
    validation_split=0.1, shuffle=True, validation_batch_size=1272, validation_freq=5
)
```
